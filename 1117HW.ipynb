```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((728, 768)),
    "Center Cropped": transforms.CenterCrop(350),
    "Color Jittered": transforms.ColorJitter(brightness=0.8, contrast=0.7, saturation=1.2),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=2),
    "Randomly Rotated": transforms.RandomRotation(75)
}

# 加載圖片
img_url = 'https://pbs.twimg.com/media/FnJvD2yagAUyx_O?format=jpg&name=4096x4096'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img)
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((1920, 768)),
    "Center Cropped": transforms.CenterCrop(360),
    "Color Jittered": transforms.ColorJitter(brightness=1.0, contrast=0.6, saturation=0.9),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=3),
    "Randomly Rotated": transforms.RandomRotation(50)
}

# 加載圖片
img_url = 'https://i1.kknews.cc/kgsGNOFizgipn57uo0NpVWjcJ05DbPbxlA/0.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img, cmap='gray')
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
